```python
import pandas as pd
import os
```

```python
green_trip_files = []
for file in os.listdir('.'):
    if file.endswith('.csv.gz'):
        green_trip_files.append(pd.read_csv(file,low_memory=False,parse_dates=['lpep_pickup_datetime','lpep_dropoff_datetime']))
```

```python
#Q1
green_trip_df = pd.concat(green_trip_files)
green_trip_df.shape
```

```python
#Q2
#if I made it compare using != 0 instead it would result in 257400 rows but it would keep the NaNs in the data
green_trip_filtered = green_trip_df[(green_trip_df['passenger_count']>0) & (green_trip_df['trip_distance']>0)]
green_trip_filtered.shape
```

```python
#Q5
[col_name for col_name in green_trip_filtered.columns if 'ID' in col_name]
```

```python
green_trip_filtered = green_trip_filtered.rename(columns = {
    'VendorID':'vendor_id'
    ,'RatecodeID':'ratecode_id'
    ,'PULocationID':'pu_location_id'
    ,'DOLocationID':'do_location_id'
} )
green_trip_filtered.columns
```
